```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns

print("Libraries imported")

#to display all the columns of data
pd.pandas.set_option('display.max_columns', None)
```

```python
#loading data and printing dimensions
airline = pd.read_csv('airline.csv')
print(airline.shape)
airline.head(20)
```

```python
airline.describe()
```

```python
airline.info()
```

```python
#cleaning column names
from skimpy import clean_columns
airline= clean_columns(airline)
airline.rename(columns= {'dayof_month':'day_of_month'},inplace = True)
```

```python
#dropping irrelevant columns and the columns providing similar information
data = airline.copy()
data = data.drop(['dot_id_reporting_airline','iata_code_reporting_airline','origin_airport_id','origin_airport_seq_id',
                 'origin_city_market_id','origin_state','origin_state_fips','origin_wac','dest_airport_id',
                  'dest_airport_seq_id','dest_city_market_id','dest_state','dest_state_fips','dest_wac'],axis=1)
```

```python
data.shape
```

```python
#changing the type of categorical variables
col = ['day_of_month','day_of_week','flight_number_reporting_airline','dep_del_15','departure_delay_groups','arr_del_15',
       'arrival_delay_groups']
for feature in col:
    data[feature]= data[feature].astype('category')
```

```python
data.info()
```

```python
chr_var = ['day_of_month','day_of_week','reporting_airline','tail_number','flight_number_reporting_airline','origin_airport',
           'origin_city_name','origin_state_name','dest_airport','dest_city_name','dest_state_name','dep_del_15',
           'departure_delay_groups','dep_time_blk','arr_del_15','arrival_delay_groups','arr_time_blk']
```

```python
#number of unique categories in each of the categorical variables
for feature in chr_var:
    print(feature, len(data[feature].unique()))
    print(data[feature].unique(),'\n')
    
'''origin airports are 351 and origin cities are 344 which means there are more than 1 airport in some of the cities.
Similarly destination airport are 350 and destination cities are 343

There is a block of 0001- 0559 in dep and arrival block columns'''
```

```python
#finding the unique categories which have appeared less than 1% for each variable
for feature in chr_var:
    temp= data[feature].value_counts()/len(data)
    print(feature, '(',len(temp[temp<0.01]),')')
    print(temp[temp<0.01])
    print('\n')
```

```python
# perentage of missing values

features_with_na = [feature for feature in data.columns if data[feature].isnull().sum()>1]
print("Percentage of missing values")
for feature in features_with_na:
    print(feature,":- ", np.round(data[feature].isnull().mean()*100,4), "%")
```

```python
#checking if there is some absurd value in dep_time due to which dep_delay is null
data[(data['dep_delay'].isnull()) & (data['dep_time'].notnull())]
```

```python
#checking if there is some absurd value in arr_time due to which arr_delay is null
data[(data['arr_delay'].isnull()) & (data['arr_time'].notnull())]
```

```python
#number of observations in each unique category of a variable
var = ['day_of_month','day_of_week','reporting_airline','dep_time_blk','arr_time_blk']
for feature in var:
    data[feature].value_counts().sort_index(axis=0, ascending= True).plot(kind='bar')
    plt.title(feature)
    plt.ylabel('count')
    plt.show()
```

```python
#top 20 categories in a variables wrt no. of observations
for feature in ['origin_airport','origin_state_name','dest_airport','dest_state_name']:
    data[feature].value_counts()[:20].plot(kind= 'bar')
    plt.title(feature)
    plt.ylabel('Count')
    plt.show()      
```

```python
num_var = ['dep_delay','taxi_out','taxi_in','arr_delay','actual_elapsed_time','air_time','distance','carrier_delay',
           'weather_delay','nas_delay','security_delay','late_aircraft_delay']
num_var
```

```python
#histogram and boxplot for numeric variables
for feature in num_var:
    plt.figure(figsize = (12,4))
    plt.subplot(1,2,1)
    data[feature].hist()
    plt.title(feature)
    plt.subplot(1,2,2)
    data.boxplot(column = feature)
    plt.title(feature)
    plt.show()
```

```python
#bar plot for counts of 0 and 1 in departure delay and arrival delays
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
data['dep_del_15'].value_counts().plot.bar()
plt.title('Departures')
plt.ylabel('Count')
plt.subplot(1,2,2)
data['arr_del_15'].value_counts().plot.bar()
plt.title('Arrivals')
plt.ylabel('Count')
plt.show()
```

```python
#percentage of delay and non delay departures wrt to top 25 origin airport and similarly for arrivals
cross_tab= pd.crosstab(index= data['origin_airport'], columns= data['dep_del_15'], normalize = 'index')
cross_tab.sort_values(by = 0.0)[:25].plot.bar()
plt.title('Origin Airport')
plt.ylabel("Percentage")
plt.show()
```

```python
cross_tab1= pd.crosstab(index= data['dest_airport'], columns= data['arr_del_15'], normalize = 'index')
cross_tab1.sort_values(by = 0.0)[:25].plot.bar()
plt.title('Destination Airport')
plt.ylabel("Percentage")
plt.show()
```
